```python
#a. What is the prices mean, median and population distribution of Asus, Samsung, 
#Apple, Xiaomi and Huawei? (use multiple boxplot)
data_brand=data[data['Brand'].isin(['Asus','Apple','Huawei','Xiaomi','Samsung'])]

data_boxplot=sns.boxplot(y='Price',x='Brand',data=data_brand)
```

###3. Exploratory Data Analysis. Apply Zscore analysis and find the outlier. How many percent (%) outlier data compares to all data? (20%)

```python
data_price = data[['Price']]
data_price
```

```python
stats.zscore(data_price, nan_policy='omit')
```

```python
data_zscore = data_price[(np.abs(stats.zscore(data_price, nan_policy='omit')) < 3).all(axis=1)]
data_zscore.shape
```

```python
data_zscore.plot(kind='box', figsize=(5,5))
plt.show()
```

```python
#Dikarenakan data_zscore.shape = (1324,1), artinya terdapat 1324 rows yang bukan outliers. 
#Sehingga total row - 1324 rows merupakan total outliers

#Total outliers = 1359-1324=35
Persentase_outliers = (35/1359)*100
print(Persentase_outliers)
```
